Generate dataset

```python
import matplotlib.pyplot as plt
import pandas as pd

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40]
height = [145, 151, 165, 173]
weight = [45, 55, 65, 75]
```

```python
health_data = pd.DataFrame({'Person': people, 'Age': age, 'Height': height, 'Weight': weight})

print(health_data)
```

Scatter Plot:

```python
plt.figure(figsize=(8, 6))
plt.scatter(health_data['Age'], health_data['Height'], color='blue', label='Height')
plt.scatter(health_data['Age'], health_data['Weight'], color='red', label='Weight')
plt.title('Health Care Data - Scatter Plot')
plt.xlabel('Age')
plt.ylabel('Height/Weight')
plt.legend()
plt.grid(True)
plt.show()
```

Explanation: The scatter plot visualizes the relationship between age and both height and weight. Each point represents a person, with their age on the x-axis and their height or weight on the y-axis. This allows us know the relationship age, height and weight.

Bar Chart

```python
plt.figure(figsize=(8, 6))
bar_width = 0.35
age_index = range(len(health_data))
plt.bar(age_index, health_data['Height'], bar_width, color='blue', label='Height')
plt.bar([i + bar_width for i in age_index], health_data['Weight'], bar_width, color='red', label='Weight')
plt.xlabel('People')
plt.ylabel('Height/Weight')
plt.title('Health Care Data - Bar Chart')
plt.xticks([i + bar_width / 2 for i in age_index], health_data['Person'])
plt.legend()
plt.show()
```

Explanation: The bar chart compares the height and weight of each person in the dataset. Each person's name is represented on the x-axis, and their corresponding height and weight are shown side by side as bars. This allows us to compare height and wight for each individual.

Histogram

```python
#for heights
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40, 45]
height = [145, 151, 165, 173]

plt.figure(figsize=(5, 3))
plt.hist(age[:-1], bins=age, weights=height, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('Height')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

```python
#for weight
import matplotlib.pyplot as plt

people = ['kiran', 'arun', 'vijay', 'varun']
age = [25, 30, 35, 40, 45]
weight=[45, 55, 65, 75]

plt.figure(figsize=(5, 3))
plt.hist(age[:-1], bins=age, weights=weight, edgecolor='black', alpha=0.7)
plt.xlabel('Age')
plt.ylabel('weight')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.show()
```

Explanation: The histogram displays the distribution of height and weight in the dataset. It shows how many people fall into each range of height and weight. This visualization helps to understand the distribution of these height and weights.
